```python
import ROOT
import numpy as np
import pandas as pd
import random
```

```python
fdata = ROOT.TFile("/home/data/AMS_Data/RisultatiDATI.root")
fmc = ROOT.TFile("/home/data/AMS_Data/RisultatiMC.root")
```

```python
def tree_to_frame(tree):
    ROOT.gROOT.SetBatch(ROOT.kTRUE)
    sers = {}
    for branch in tree.GetListOfBranches():
        bName = branch.GetName()
        tree.Draw(bName)
        size = tree.GetSelectedRows()
        buf = tree.GetV1()
        buf.SetSize(size)
        sers[bName] = pd.Series(np.array(buf,copy=True))
    return pd.DataFrame(sers)
```

```python
list(fdata.GetListOfKeys())
```

```python
pdat = tree_to_frame(fdata.Get('grandezzequal'))
pmc  = tree_to_frame(fmc.Get('grandezzequal'))
```

$\beta^2 = \frac{1}{1+m^2/R^2}$

```python
fr = pdat
figsize(20,8)
for fr, cl in [(pdat, 'k'), (pmc, 'r')]:
    scatter(fr['R'].values, fr['Beta'].values, s=2, c=cl, lw=0, marker=',')
xlim(0,10)
ylim(0,1)

R = np.linspace(0,10,100)

for m in np.linspace(0.5,3,10):  
    plot(R, np.sqrt(1/(1+(m/R)**2)), c='b') 
```

### Adding masses to frames

```python
for fr in [pdat, pmc]:
    fr['CalcMass'] = fr['R']*np.sqrt(1-fr['Beta']**2)/fr['Beta']
```

```python
figsize(20,8)
for fr, cl in [(pdat, 'k'), (pmc, 'r')]:
    fr = fr[(fr['R'] < 10) & (fr['CalcMass'] > 1)]
    scatter(fr['CalcMass'].values, fr['Beta'].values, s=3, c=cl, lw=0, marker=',')
xlim(0,10)
ylim(0,1)
```

```python
import ROOT
import numpy as np
import pandas as pd
import random
```

```python
fdata = ROOT.TFile("/home/data/AMS_Data/RisultatiDATI.root")
fmc = ROOT.TFile("/home/data/AMS_Data/RisultatiMC.root")
```

```python
treedat = fdata.Get("grandezzequal")
treemc  = fmc.Get("grandezzequal")
```

```python
def tree_to_frame(tree):
    ROOT.gROOT.SetBatch(ROOT.kTRUE)
    sers = {}
    for branch in tree.GetListOfBranches():
        bName = branch.GetName()
        tree.Draw(bName)
        size = tree.GetSelectedRows()
        buf = tree.GetV1()
        buf.SetSize(size)
        sers[bName] = pd.Series(np.array(buf,copy=True))
    return pd.DataFrame(sers)
```

```python
datFrame = tree_to_frame(treedat)
mcFrame = tree_to_frame(treemc)
```
